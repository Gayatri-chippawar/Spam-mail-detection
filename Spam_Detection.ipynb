**Importing Dependancies**

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
mail = pd.read_csv('../datasets/mail_data.csv')
```

```python
mail.head()
```

```python
mail['Category'].value_counts()
```

**Filtering the data**

```python
mail_data = mail.where((pd.notnull(mail)),'')
```

```python
mail_data.head(2)
```

**Vectorizing Manually**

```python
mail_data.loc[mail_data['Category'] == 'spam','Category'] = 0
mail_data.loc[mail_data['Category'] == 'ham','Category'] = 1
```

```python
mail_data.head()
```

**Dividing the Data**

```python
x = mail_data['Message']
y = mail_data['Category']
```

```python
print(x)
```

```python
print(y)
```

**Spliting in train-test**

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=4)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

```python
print(y.shape,y_train.shape,y_test.shape)
```

**Vectirization**

```python
feature = TfidfVectorizer(min_df=1,stop_words='english',lowercase=True)
```

```python
x_train_feature = feature.fit_transform(x_train)
```

```python
x_test_feature = feature.transform(x_test)
```

```python
print(x_train_feature)
```

```python
print(x_test_feature)
```

**Importing model**

```python
model = LogisticRegression()
```

```python
Y_train = y_train.astype('int')
Y_test = y_test.astype('int')
```

```python
model.fit(x_train_feature,Y_train)
```

```python
model_predict = model.predict(x_train_feature)
train_accuracy = accuracy_score(Y_train,model_predict)
```

```python
print(train_accuracy)
```

```python
model_predict1 = model.predict(x_test_feature)
test_accuracy = accuracy_score(Y_test,model_predict1)
```

```python
print(test_accuracy)
```
